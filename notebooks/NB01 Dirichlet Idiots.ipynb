```python
import pandas as pd
import re

from wordcloud import WordCloud

import matplotlib.pyplot as plt
# %pylab inline
%matplotlib inline

plt.rcParams["figure.figsize"] = (10,10)
```

```python
jayDat = pd.read_csv('jayavel_GetItForMeData.csv')
jayDat.head(4) 
```

```python
titles = jayDat['Title'].tolist()

titles = [re.sub('[^A-Za-z0-9]+', ' ', title) for title in titles]
titlesJoined = " ".join(titles)
wordList = re.sub("[^\w]", " ",  titlesJoined).split()
wordList[:5]
```

```python
wordcloud = WordCloud(width=3000,height=3000,background_color='white').generate(titlesJoined) 
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
# plt.savefig('jayDat_wordCloud.png')
```
